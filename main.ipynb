```python
#imports
import pandas as pd
import requests
from dotenv import load_dotenv
import os

load_dotenv()  # Loads .env variables
```

```python
all_apps = pd.read_csv("data/all_apps.csv")
#print(f"loaded {all_apps} apps.")
```

```python
# store.steampowered.com/api/appdetails
appid = 10
url = "https://store.steampowered.com/api/appdetails/APPID"

params = {
    "appids": appid
}

response = requests.get(url, params=params)
response.raise_for_status()

data = pd.DataFrame(response.json())

os.makedirs("data/appdetails", exist_ok=True)
#data.to_csv(f"data/appdetails/{appid}.csv", index=False)
data.to_json(f"data/appdetails/{appid}.json", orient="records", indent=2)
```

```python
# webscraping
appid = 2945940
url = f"https://store.steampowered.com/app/{appid}/?l=english"

response = requests.get(url+str(appid))
response.raise_for_status()

os.makedirs("data/tags", exist_ok=True)
with open(f"data/tags/{appid}.html", "w", encoding="utf-8") as f:
    f.write(response.text)
```

```python
# api.steampowered.com/IStoreBrowseService/GetItems
appid = 2945940
url = f"https://api.steampowered.com/IStoreBrowseService/GetItems/v1"

params = {
    "input_json": '{"ids":[{"appid":2945940}],"context":{"language":"english","country_code":"US","steam_realm":1},"data_request":{"include_assets":true,"include_release":true,"include_platforms":true,"include_all_purchase_options":true,"include_screenshots":true,"include_trailers":true,"include_ratings":true,"include_tag_count":10000,"include_reviews":true,"include_basic_info":true,"include_supported_languages":true,"include_full_description":true,"include_included_items":true,"include_assets_without_overrides":true,"include_links":true}}'
}

response = requests.get(url, params=params)
response.raise_for_status()

os.makedirs("data/GetItems", exist_ok=True)
with open(f"data/GetItems/{appid}.json", "w", encoding="utf-8") as f:
    f.write(response.text)
```
