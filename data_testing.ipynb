```python
#imports
import pandas as pd
import requests
from dotenv import load_dotenv
import os
import json

load_dotenv()  # Loads .env variables
```

```python
DIRECTORY = "data/storebrowse_items"

irregular = []

def check_file(data):
    store_items = data.get("response", {}).get("store_items", [])
    for item in store_items:
        check_app(item)

def check_app(item):
    # <<<<<<<<<<<<<<<<<<<<<<
    if item.get("type") != 0 and item.get("type") != None:
        irregular.append({
            "appid": item.get("appid"),
            "name": item.get("name"),
            "type": item.get("type"),
        })

counter = 0
for filename in os.listdir(DIRECTORY):
    if len(irregular) >= 100:
        break

    if filename.endswith(".json"):  # Check if file is a JSON file
        counter += 1
        print(f"checking {filename} ({counter / 1483 * 100:.2f} %)...", end="\r")
        file_path = os.path.join(DIRECTORY, filename)
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)  # Load JSON data
            check_file(data)
print("")
print(f"irregular ({len(irregular)}) = {irregular}")
```
